```python
# Repository Cleanup and Production Refactor

This notebook inspects the repository structure, identifies obsolete files, refactors the production-ready configuration, and removes stale artifacts.
<VSCode.Cell id="#VSC-e20853e7" language="markdown">
## 1. Inspect Repository Files

List the top-level repository contents and identify duplicate or stale frontend code folders.
```

```python
import os
from pathlib import Path

root = Path('/workspaces/FamilyOps_AI')
entries = sorted([p.name for p in root.iterdir()])
print('Top-level entries:')
for name in entries:
    print('-', name)

frontend_exists = root.joinpath('frontend').exists()
src_exists = root.joinpath('src').exists()
print('\nfrontend exists:', frontend_exists)
print('src exists:', src_exists)
print('backend exists:', root.joinpath('backend').exists())
```

## 2. Identify Unwanted Files

The repository contains an active root Next.js app under `src/`, while `frontend/` contains a duplicate copy. The following files are stale or unnecessary for the production layout:
- `frontend/`
- `main.py`
- `pyproject.toml`
- `sql query`

The production-ready structure should keep only the root frontend source in `src/` and the backend code in `backend/`.

## 3. Define Production Folder Structure

Use the following production-ready layout:

- `/src/` — Next.js frontend app router source
- `/backend/` — FastAPI backend and AI service code
- `/.gitignore` — ignore build, dependency, and temp artifacts
- `/README.md` — project overview and folder layout
- `/package.json` — frontend package scripts and dependencies
- `/backend/requirements.txt` — backend Python dependencies

## 4. Refactor Code into Production Modules

Update the production configuration for the frontend and backend so the repo is ready for deployment.

```python
from pathlib import Path

root = Path('/workspaces/FamilyOps_AI')

# Update frontend rewrite config to use BACKEND_URL when available
next_config = root.joinpath('next.config.js')
next_config.write_text(
'''
const backendUrl = process.env.BACKEND_URL || 'http://localhost:8000';

/** @type {import('next').NextConfig} */
const nextConfig = {
  reactStrictMode: true,
  async rewrites() {
    return [
      {
        source: '/api/v1/:path*',
        destination: `${backendUrl}/api/v1/:path*`,
      },
    ];
  },
};

module.exports = nextConfig;
'''
)

# Update backend runner for production host and port configuration
backend_run = root.joinpath('backend', 'run.py')
backend_run.write_text(
'''
from dotenv import load_dotenv
from pathlib import Path
import os

env_path = Path(__file__).parent / '.env'
load_dotenv(env_path, override=False)

host = os.getenv('HOST', '0.0.0.0')
port = int(os.getenv('PORT', '8000'))

import uvicorn
from app.core.config import settings

if __name__ == '__main__':
    uvicorn.run(
        'main:app',
        host=host,
        port=port,
        reload=settings.debug,
        log_level='debug' if settings.debug else 'info',
    )
'''
)

# Add cleanup guidance to README for production folder structure
readme = root.joinpath('README.md')
readme_text = readme.read_text()
updated_readme = readme_text + '

## Production folder structure
' + 
,
,
,
,
,
,
,
,
,
,
```
